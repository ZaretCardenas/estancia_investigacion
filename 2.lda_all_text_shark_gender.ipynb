```python
from nltk.stem import WordNetLemmatizer, SnowballStemmer
import contractions

import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import contractions

nltk.download('stopwords')
nltk.download('wordnet')

def preprocess(text):
    """This function applies NLP preprocessing on a single string.
       These steps include converting text to lowercase, expanding contractions, removing punctuation,
       removing numbers, removing stopwords, lemmatizing, and removing short words."""
    
    # convert to lowercase
    text = ' '.join([w.lower() for w in text.split()])
    
    # expand contractions  
    text = ' '.join([contractions.fix(word) for word in text.split()])
    
    text = re.sub(r'\b(knife|kayak)\b', 'product', text)
    text = re.sub(r'\b(therapist)\b', 'service', text)

    pattern = r'\b(?:t(?:herapist|heraphist)|th(?:erapist|erapists|eraphist|eraphists))\b'
    
    # Replace matches with the word 'product'
    text = re.sub(pattern, 'service', text, flags=re.IGNORECASE)
    
    # remove punctuation
    text = ''.join([i for i in text if i not in string.punctuation])
    
    # remove numbers
    text = ' '.join(re.sub("[^a-zA-Z]+", " ", text).split())

    # remove stopwords
    stopwords_list = [sw for sw in nltk.corpus.stopwords.words('english') if sw not in ['not', 'no']]
    text = ' '.join([w for w in text.split() if w not in stopwords_list])

    # lemmatization
    lemmatizer = WordNetLemmatizer()
    text = ' '.join([lemmatizer.lemmatize(w) for w in text.split()])

    # remove short words (words with fewer than 3 characters)
    text = ' '.join([w.strip() for w in text.split() if len(w.strip()) > 3])

    return text
```

```python
import os
import pickle
import gensim
from gensim import corpora

def lda_models_all_text(text_data_all_h,text_data_all_m, num_topics_h,num_topics_m):
    root = f'6_exp_lda_all_text_by_shark_gender/lda/'
    os.makedirs(root, exist_ok=True)  # Create directory if it doesn't exist
    print( 'original h',text_data_all_h)
    print( 'original m',text_data_all_m)
    # Helper function to process LDA for a given dataset
    def process_lda(text_data, root_prefix,num_topics):
        # Preprocess the text data
        text_data_processed = [clean_and_lemmatize2(phrase) for phrase in text_data]
        print( 'process',text_data_processed)
        #tokenized_text = [phrase.split() for phrase in text_data_processed]  # Tokenize text data

   
        text_data_processed = [preprocess(phrase) for phrase in text_data_processed ]
        print( 'process2',text_data_processed)

        tokenized_text = list(sent_to_words(text_data_processed ))

        
        tokenized_text=remove_stopwords(tokenized_text)
        print( 'tesxto tokenizado',tokenized_text)
        # Create dictionary and corpus
        dictionary = corpora.Dictionary(tokenized_text)

        corpus = [dictionary.doc2bow(text) for text in tokenized_text]

        # Save dictionary and corpus
        pickle.dump(corpus, open(f'{root}{root_prefix}_corpus.pkl', 'wb'))

        dictionary.save(f'{root}{root_prefix}_dictionary.gensim')

        # Train and save the LDA model
        ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=20)

        ldamodel.save(f'{root}{root_prefix}_model.gensim')

        return tokenized_text



    # Process the LDA for men (text_data_all_h) and women (text_data_all_m)
    tokenized_text_h=process_lda(text_data_all_h, 'h_key',num_topics=num_topics_h)  # For men
    tokenized_text_m=process_lda(text_data_all_m, 'm_key',num_topics=num_topics_m)  # For women
    
    return tokenized_text_h,tokenized_text_m

```

```python
import re

def find_texts_with_therapist(texts):
    """This function finds and returns all texts that contain the word 'therapist' or 'therapists'."""
    # Define the regex pattern to match 'therapist' or 'therapists'
    pattern = r'\bsofa?\b'
    
    # Filter and return texts containing the word 'therapist' or 'therapists'
    return [text for text in texts if re.search(pattern, text, re.IGNORECASE)]

def count_therapist_occurrences(texts):
    """This function counts how many times the word 'therapist' or its variations appear in the list of texts."""
    # Define the regex pattern to match 'therapist' or 'therapists'
    pattern = r'\bsof?\b'
    
    # Initialize a counter
    count = 0
    
    # Count occurrences of 'therapist' or 'therapists' in each text
    for text in texts:
        matches = re.findall(pattern, text, re.IGNORECASE)
        count += len(matches)
    
    return count

# Example usage:
texts = [
    "I went to see my therapist today.",
    "This is a random sentence without the word.",
    "Therapists are important for mental health.",
    "No therapists here."
]

text_data_processed_m = [preprocess(phrase) for phrase in text_data_all_m]



matching_texts = find_texts_with_therapist(text_data_processed_m)
print(matching_texts)

count = count_therapist_occurrences(matching_texts )
print(count)
```

```python
def get_topics_words(lda_model, feature_names, n_top_words=5):
    topics_words = {}
    for idx, topic in enumerate(lda_model.components_):
        # Get the top n words for each topic
        top_words_idx = topic.argsort()[-n_top_words:][::-1]  # Get indices of top words
        top_words = [feature_names[i] for i in top_words_idx]  # Get top words
        topics_words[f"Topic {idx + 1}"] = top_words  # Store in a dictionary
    return topics_words

# Get topics and their words
topics = get_topics_words(lda, vectorizer.get_feature_names_out(), n_top_words=5)

# Print the topics and their words
for topic, words in topics.items():
    print(f"{topic}: {', '.join(words)}")
```

```python


import pandas as pd
import re

# Lista de archivos de hombres y mujeres

# Inicializar listas para almacenar los resultados
genders = []
chapters = []
pitches = []

# Regex para identificar los patrones
pattern_spanish_men = r'Capitulo(\d+)pitch(\d+)_H_lab.txt'
pattern_spanish_women = r'Capitulo(\d+)pitch(\d+)_M_lab.txt'
pattern_spanish_mixed = r'Capitulo(\d+)pitch(\d+)_B_lab.txt'

pattern_english_men = r'Men_Chapter(\d+)_Pitch(\d+)_Season8.txt'
pattern_english_women = r'Women_Chapter(\d+)_Pitch(\d+)_Season8.txt'
pattern_english_mixed = r'Mixed_Chapter(\d+)_Pitch(\d+)_Season8.txt'

# Procesar archivos de hombres
for filename in total_files_H:
    match_spanish_men = re.match(pattern_spanish_men, filename)
    match_english_men = re.match(pattern_english_men, filename)
    
    if match_spanish_men:
        genders.append('male')
        chapters.append(int(match_spanish_men.group(1)))
        pitches.append(int(match_spanish_men.group(2)))
    elif match_english_men:
        genders.append('male')
        chapters.append(int(match_english_men.group(1)))
        pitches.append(int(match_english_men.group(2)))

# Procesar archivos de mujeres y mixtos
for filename in total_files_M:
    match_spanish_women = re.match(pattern_spanish_women, filename)
    match_spanish_mixed = re.match(pattern_spanish_mixed, filename)
    match_english_women = re.match(pattern_english_women, filename)
    match_english_mixed = re.match(pattern_english_mixed, filename)
    
    if match_spanish_women:
        genders.append('female')
        chapters.append(int(match_spanish_women.group(1)))
        pitches.append(int(match_spanish_women.group(2)))
    elif match_spanish_mixed:
        genders.append('mixto')
        chapters.append(int(match_spanish_mixed.group(1)))
        pitches.append(int(match_spanish_mixed.group(2)))
    elif match_english_women:
        genders.append('female')
        chapters.append(int(match_english_women.group(1)))
        pitches.append(int(match_english_women.group(2)))
    elif match_english_mixed:
        genders.append('mixto')
        chapters.append(int(match_english_mixed.group(1)))
        pitches.append(int(match_english_mixed.group(2)))

# Crear el DataFrame final
df = pd.DataFrame({
    'Gender': genders,
    'Chapter': chapters,
    'Pitch': pitches
})
df['key']=df['Chapter'].astype('str')+'_'+df['Pitch'].astype('str')
# Mostrar el DataFrame
df.to_csv('gender_by_txt.csv',index=False)
```
